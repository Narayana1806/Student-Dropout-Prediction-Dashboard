```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Data_Set.csv")

# Get feature names (columns)
features = df.columns

# Equal values for each feature (only names matter)
values = [1] * len(features)

# Plot pie chart
plt.figure(figsize=(8,8))
wedges, texts, autotexts = plt.pie(
    values,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Paired.colors
)


# Add legend (small font, outside chart)
plt.legend(
    wedges,
    features,
    title="Features",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize="small"
)

plt.title("Features in Dataset")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Data_Set.csv")

# Count values of Dropped_Out column
drop_counts = df["Dropped_Out"].value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
wedges, texts, autotexts = plt.pie(
    drop_counts,
    labels=drop_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Set2.colors
)

# Add legend (small size, outside chart)
plt.legend(
    wedges,
    drop_counts.index,
    title="Dropped Out Status",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize="small", labels=["Yes","No"]
)

plt.title("Dropped Out Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Data_Set.csv")

# Calculate number of unique values for each feature
unique_counts = df.nunique()

# Plot bar chart
plt.figure(figsize=(10,6))
unique_counts.plot(kind="bar", color="teal")

plt.title("Unique Value Counts for Each Feature")
plt.xlabel("Features")
plt.ylabel("Number of Unique Values")

plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/Data_Set.csv")

# Get feature names (columns)
features = df.columns

# Assign equal values (or you can replace with actual counts if needed)
values = [1] * len(features)

# Plot pie chart with percentages
plt.figure(figsize=(8,8))
wedges, texts, autotexts = plt.pie(
    values, startangle=90, colors=plt.cm.Paired.colors,
    autopct='%1.1f%%', textprops={'fontsize': 8}
)

# Add legend on the LEFT side
plt.legend(
    wedges, features, title="Features",
    loc="center left",
    bbox_to_anchor=(-0.3, 0.5),
    fontsize=8
)

plt.title("Features in Dataset")
plt.show()
```
